## The Foursquare API Data:

We will need data about different venues in various neighborhoods of that particular borough. 
In order to obtain that information we will use "Foursquare" locational information. Foursquare is a location data provider with information about all manner of venues and events within an area of interest. Such information includes venue names, locations, menus and even photos. As such, the foursquare location platform will be used as the sole data source since all the stated required information can be obtained through the API. 

After finding the list of neighborhoods, we then relate to the Foursquare API to gather information about venues inside each and every neighborhood. For each neighborhood, we have chosen the radius to be 100 meter.

The data retrieved from Foursquare contained information of venues within a specified distance of the longitude and latitude of the postcodes. The information obtained per venue as follows:

    I. Neighborhood
    II. Neighborhood Latitude
    III. Neighborhood Longitude
    IV. Venue
    V. Name of the venue e.g. the name of a store or restaurant
    VI. Venue Latitude
    VII. Venue Longitude
    VIII. Venue Category
